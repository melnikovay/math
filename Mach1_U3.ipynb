Почему прямые не кажутся перпендикулярными? 

Так как данные на графике представлены в разных масштабах по оси Х и У. Если показать в одном масштабе, 
то прямые будут перпендикулярными

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-5, 5, 5)
y = 3*x+1
y1 = -1/3*x+1
plt.plot(x,y)
plt.plot(x,y1)
plt.xlim (-15, 15)
```

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
x =[]
y =[]
y2=[]
r=2
for i in range (-10,11,1):
    x_= i/5
    x.append(x_)
    y.append(((r**2)-(x_**2))**0.5)
    y2.append(-((r**2)-(x_**2))**0.5)

plt.plot(x,y)
plt.plot(x,y2)
plt.ylim (-3, 3)
plt.xlim (-4.4, 4.4)
print(x)
print(y)
```

```python
x =[]
y =[]
y1=[]
a=3
b=2
for i in range (-15,16,1):
    x_= i/5
    x.append(x_)
    y.append(((b**2)*(1-((x_**2)/(a**2))))**0.5)
    y1.append(-(((b**2)*(1-((x_**2)/(a**2))))**0.5))

plt.plot(x,y)
plt.plot(x,y1)
plt.ylim (-3, 3)
plt.xlim (-4.4, 4.4)
print(x)
print(y)
```

```python
x=[]
y=[]
y2=[]
a=2
b=3
for i in range (-15,16,1):
    x_= i/5
    x.append(x_)
    y.append(((((b**2)*(x_**2))/(a**2))-(b**2))**0.5)
    y2.append(-(((((b**2)*(x_**2))/(a**2))-(b**2))**0.5))
print(x)
plt.plot(x,y)
plt.plot(x,y2)
```
